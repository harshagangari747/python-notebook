#### Multiple plots
* `color` - gives the graph a color. Use full name or shortcuts. Example red/r, blue/b, green/g, black/k, yellow/y
* `marker` - symbol used to represent an individual datapoint on the graph. Available markers, 'o, s, ^, v, > ,<, D, P, *, X
* `linestyle` - Style of line used to connect datapoints. Example: **'--' , '---' ,'-' , ':'**

```python
mp.plot(vals,np.sin(vals),label="sin(x)",color='g',linestyle=':')
mp.plot(vals,np.cos(vals),label="cos(x)",color='b',marker='o',linestyle=None)    ## k for black
mp.legend()    ## you must call legend() to show up the labels on the plot
mp.xlabel("x-axis")
mp.ylabel('$y-axis$')    ## $ text$ - makes into italics
```

### Bar and histogram graphs
* use `bar()` function to plot bar graphs
* `barh()` - For horizontal bar plot
* `color`(attr) - To change the colors of the plot
* `edgecolor` (attr) - To give borde for each bar
* `align` (attr) - Centers the plot. Default it **edge**. Available : **edge** and **center**
* `hatch` (attr) - Gives a pattern to the bar. Ex: *, |, -, x
* `width` (attr) - Set the width (thickness) of each bar. Accepts floating point numbers
* `bottom` (attr) = Set the bottom height to a number corresponding to the y-axis. Ex: If set to 5, then all the bars start from the line y=5 on y-axis

* #### `hist()` - Use this function for histogram
* `range()` - Takes a tuple of two values that specified the range of values on x-axis to display
* `bins` (attr) = Specify the intervals or ranges into which data is divided. Set to `auto` to leave the decision for matplotlib
* `histtype` (attr) - Determines the style of the histogram. Available: **bar**, **barstacked**, **step**, **stepfilled**
* `cumulative` (attr) - Displays the cumulative count of values rather than just count in each bin. Availabe: **True**, **False**, **1**, **-1** (to start from opposite direction)
* `density` formerly **normed** (attr) - When set to ***True***, the histogram displays probability density function (PDF). Default is ***False***

#### Images and Contours

#### Making multiple panel plots
* #### subplot() and subplots()
* `subplot()` - Use this to plot multiple plots in a single plot.
   * Takes three agurments *arg1*: **number of rows**, *arg2*:** number of columns**, *arg3*:**slot number to which the current plot belongs**
      * Instead of three args, send them as an integer. Example instead of using mp.subplot(2,1,2), use mp.subplot(212)
* `subplots()` - Use this when you want to plot subplots programmatically

```python
mp.subplot(2,2,1)    #### specifies plot with 1 row, 2 cols, and belong to 1st slot
mp.plot(xvalues,yvalues)
mp.xlabel('sinx')
mp.subplot(2,2,2)    #### specifies plot with 1 row, 2 cols, and belong to 2nd slot
mp.plot(xvalues,np.cos(xvalues))
mp.xlabel('cos')
mp.ylabel('Y')


mp.subplot(223)    #### specifies plot with 1 row, 2 cols, and belong to 1st slot
mp.plot(xvalues,np.arcsinh(xvalues))
mp.subplot(223)    #### specifies plot with 1 row, 2 cols, and belong to 2nd slot
mp.plot(xvalues,np.tanh(xvalues))
```

* `mp.clf()` - Clears the figure
* `cla()` - clears an axes. Call on the axes object

#### Working with transformations
* Matplotlib works with `4 Coordinate systems`
   * Data
   * Axes
   * Figure
   * Display
* `transData` - Transforms the datapoint to display. Returns pixel position of the data point with respect to the axes
* `transAxes` - Transforms the coordinates with in the axes frame to coordinates with in the display frame
* `transFigure` - Transforms the coordinates with in the Figure to coordinates with in the display frame
* `inverted()` - Transforms the coordinates with in the display to coordinates of the data point

```python
tfigure = mp.figure(figsize=[2,2])
ax1 = tfigure.add_axes([0,0,1,1])
ax1.plot(1,1,'ro')
print(ax1.transData.transform((1,1)))
```

```python
tfigure = mp.figure(figsize=[2,2])
ax1 = tfigure.add_axes([0,0,1,1])
ax1.plot(1,1,'ro')
print(ax1.transAxes.transform((1,1)))
```

```python
tfigure = mp.figure(figsize=[2,2])
ax1 = tfigure.add_axes([0,0,1,1])
ax1.plot(1,1,'ro')
ax1.transData.inverted(1)
```

#### Controlling Axes and Ticks
* Axes label, Ticks and spines
   * `yaxis.grid([True|False])` - Use this to display the grid lines across the y-axis. Call on a axes object
   * `xaxis.grid([True|False])` - Use this to display the grid lines across the x-axis. Call on a axes object
   * `yaxis.set_ticks_position` - Sets the number line on the **right** or **left** side of the graph
   * `yaxis.get_ticklocs()` - Get an array of ticks along the y-axis
   * `xaxis.get_ticklocs()` - Get an array of ticks along the x-axis
   * `[xaxis|yaxis].set_minor_locator` - Set the number of minor ticks to be displayed on the corresponding axis
   * `[xaxis|yaxis].set_major_locator` - Set the number of major ticks to be displayed on the corresponding axis
   * `set_[x|y]label('axis name here')` - Sets the label for the corresponding axis
   * `[x|y]axis.set_label_text('new label text here')` - Alternate method for setting label
   * `[x|y]axis.set_label_position(['left'|'right'])` - Places the label of the corresponding axis on the graph. left/right for yaxis and top/bottom for xaxis
ax.yaxis.set_label_coords(x=1.5,y=0.5)

```python
newfig = mp.figure(figsize=[3,3])
axis = newfig.add_axes([0,0,1,1])
axis.yaxis.grid(True)
axis.xaxis.grid(True)

axis.yaxis.set_ticks_position('right')
axis.xaxis.set_ticks_position('top')

axis.yaxis.get_ticklocs()
```

```python
from matplotlib import ticker
```

```python
tickfig = mp.figure(figsize=[3,3])
ax = tickfig.add_axes([0,0,1,1])
ax.xaxis.set_minor_locator(ticker.MaxNLocator(20))
ax.yaxis.set_minor_locator(ticker.MaxNLocator(40))

ax.xaxis.set_major_locator(ticker.MaxNLocator(10))
ax.yaxis.set_major_locator(ticker.MaxNLocator(5))

ax.set_ylabel('y-axis')
ax.yaxis.set_label_text('new y label')
ax.yaxis.set_label_position('right')
ax.yaxis.set_label_coords(x=1.005,y=0.5)

ax.xaxis.set_label_text('new x label')
ax.xaxis.set_label_position('bottom')
```
