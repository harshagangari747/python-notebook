```python
                                                            #Lists
series = ['got','breakingbad','vikings','house of cards','Family guy']
print("1. ",series)
print(f"2. third series: {series[2]}")
print(f"3. Length of the series list: {len(series)}")

movies = [['bahubali','dookudu','godavari'],['inception','harrypotter','badland'],['manjumeel boys']]
print(f"4. A movie suggestion for you: {movies[0][2]}")

## Add to a list in the end - append()
series.append('squid game')
print("5. ",series)

## Remove elements from a list
movies.remove(['manjumeel boys'])
print("6. ",movies)

## Sum of all elements in a list. Elements must be of type int or float type
scores = [32,66.2,88,90.32,53]
total_score = sum(scores)
print("7. ",total_score)
print("8. ",len(movies))


## Joining a list
joinedserieslist = ",".join(series)
print(f"9. My favorite series list: {joinedserieslist}")

## List comprehension
nums = [x for x in range(1,20)]
print(f'10. {nums}')

oddnums = [y for y in nums if y%2 != 0]
print(f"11. Odd nums: {oddnums}")

## Zip function
listA = ['a','e','i','o','u']
listB = [2,3,5,3,7,8] 
zippedlist = list(zip(listA,listB))
print("12. ",zippedlist)
```

```python
                                                                    #Sets
# Doesn't hold order. Doesn't store any duplicates
fruits = {'apple','mango','banana'}
print(f"1. fruits set {fruits}")

## Add to a set
fruits.add('strawberry')
print(f"2. new fruits set: {fruits}")

## Adding duplicates to set
fruits.add('mango')                              #No error but it doesn't store in the set
print(f"3. fruit basked didn't change: {fruits}")

## Remove element from a set

fruits.remove('apple')
print(f"4. fruit basket updated: {fruits}")

newfruitbasket = {'mango','raspberry','orange'}

## Difference of two sets
diff_fruits_in_fruits_set = fruits.difference(newfruitbasket)
diff_fruits_in_fruit_basket = newfruitbasket.difference(fruits)
print(f"5. fruits set have {diff_fruits_in_fruits_set} fruits different from fruits basket")
print(f"6. fruitbasket have {diff_fruits_in_fruit_basket} fruits different from fruits set")

## difference of two sets that are not in BOTH
not_in_both_baskets = fruits.symmetric_difference(newfruitbasket)      # can use interchangeably. Ex. newfruitbasket.symmetric_difference(fruits)
print(f"7. fruits that are not in BOTH sets {not_in_both_baskets}")

##Intersection of two sets
commonfruits = fruits.intersection(newfruitbasket)
print(f'8. common fruits in both baskets {commonfruits}')

## Union of two sets
unionoffruits = newfruitbasket.union(fruits)
print(f'9. whole basket {unionoffruits}')

## Sum of all elements in a set
scores = {32,66.2,88,90.32,53}
total_score = sum(scores)
print("10.", total_score)

## Length of a set
print("11.",len(newfruitbasket))


## Set comprehension
commonfruit = {f.upper() for f in fruits.intersection(newfruitbasket)}
print("12.",commonfruit)

## Zip function
setA = {1,2,4}
setB = {4,3,5}
zippedset = set(zip(setA,setB))
print("13.",zippedset)
```

```python
#simple function
def printGreeting():
    print('1. Hello')

# function with arguments
def greetUser(name,msg):
    print('2. Hello %s. Message: %s'%(name,msg))

# return a value from a function
def isEven(x):
    if x%2 == 0:
        return True
    return False

# A function is executed only when it is called and not when it is created.
# call a function using function name and provide necessary arguments if any
printGreeting()
greetUser('3. Peter','Shutup Meg!')
isNumEven = isEven(459223)
print('4.',isNumEven)

## Default values for a parameter
## 1. In the function signature, the default parameters should be at last
## 2.The parameter following default parameter must be a default parameter
## 3.While calling a function using named parameters, the parameter used with
## name should be followed by parameter that must be used with name
## 4.Calling a parameter with name can be preceeded by parameters with no name!
searchengines=('www.google.com','www.yahoo.com','www.bing.com')
def searchresults(website,query='carphotos'): ## default arguments must be at last
    if website in searchengines:
        return website+("/"+query)
    return 'Search engine not supported!'
results = searchresults("www.google.com","instagram")
print('5.',results)
print('6.',searchresults('www.yahooo.com'))
print('7.',searchresults(website='www.bing.com',query='kungfu-panda'))

## Lambda functions - very short functions
quotient = lambda x,y: x%y
print("8. ",quotient(4,5))

## Higher order functions - pass function as a parameter
listC = ["instagram","snapchat",'facebook']
def searchlist(item):
    if item in listC:
        return item
    return None

def browser(query,searchlistfunc):
    if searchlistfunc(query):
        return 'found'
    return 'Query not supported'

result = browser('snapchat',searchlist)
print("9.",result)
print("10.",browser('whatsapp',searchlist))


```

```python
import numpy as np
height = [1.87,  1.87, 1.82, 1.91, 1.90, 1.85]
weight = [81.65, 97.52, 95.25, 92.98, 86.18, 88.45]
                                                        #create numpy arrays from lists
height_array = np.array(height)
weight_array = np.array(weight)

print("==>",weight,weight_array)
print("==>",type(height_array))


## Creating n-dimensional arrays
heading("CREATING N DIMENSIONAL ARRAYS USING NUMPY")
zerodimarr = np.array(7468788)
print("==>Zero dim array",zerodimarr)

onedimarr = np.array(weight)                                                           # one dimensional array

twodimarr = np.array([[2,False,5.5,134],[6,'2',3,3]])                                  # setting any element to str makes every element as str
                                                                                       # similarly applies for float. If a str, float, int are present, 
                                                                                       # then str takes priority

threedimarr = np.array([[[3,5,6],[False,True,False]],[[3.2,6.99,49.4],['a',"bc","xyz"]]])
print("==>",twodimarr)
print("==>",type(twodimarr[0][0]))
print("==> dimension of zerodimarr: ",zerodimarr.ndim)

                                                        ## Create array of n dimension
fivedimarr = np.array(76,ndmin=5)
print("==>array with ndmin = 5 ", fivedimarr)
print("==>Three dim array",threedimarr)

heading("INDEXING ARRAYS USING NUMPY")
print("==>0,1,2 dims of threedimarr: threedimarr[0][1][2]",threedimarr[0][1][2])
print("==>second row of twodimarr: twodimarr[1]",twodimarr[1])
print("==>Another notation to get element at 1,2 pos of twodimarr[1][2]: ",twodimarr[1,2])
print("==>Same with threedimarr: threedimarr[1,1,1]", threedimarr[1,1,1])

heading("SLICING ARRAYS")
sliceOneDimArr = onedimarr[3:5]                                                       # takes all elements from index 3 till 4 (5-1)
print("==>",sliceOneDimArr)

sliceTwoDimArr = twodimarr[0:1]                                                       # takes all elements from index 0 till 0 (1-1) 
                                                                                      # and returns 2 dim array
print("==>",sliceTwoDimArr.ndim)

sliceOneDimArrWhole = onedimarr[1:5]                                                  # similar to copying all the elements of the array 
                                                                                      # to another array
sliceOneDimArrWhole[0] = 3                                                            # Changing this element will modify the elements in original array
print("==>",sliceOneDimArrWhole," originial one dim arr",onedimarr)

recallInSteps = height_array[::2]
print("==> recall height_array[::2]",recallInSteps)

recallTwoDimArr = twodimarr[0,0:2]
print("==> recall twodimarr[0,0:2]",recallTwoDimArr)

recallTwoDimArr = twodimarr[0:2,1]
print("==> recall twodimarr[0:2,1]",recallTwoDimArr)

recallTwoDimArr = twodimarr[0:1,0:3]
print("==> recall twodimarr[0:1,0:3]",recallTwoDimArr)

heading("VIEW AND COPY")
                                                        ## Copied array is individual of orginial array.
org_arr = np.array([1,2,3,4,5])
copied_arr = np.copy(org_arr)
org_arr[0] = -1                                                                    # Does affect the original array
print("==> org_arr: ",org_arr," copied array:",copied_arr)
copied_arr[0] = 10                                                                 # Value changes only in copied array and 
                                                                                   # not in orginal array
print("==> org_arr: ",org_arr," copied array:",copied_arr)

                                                        ## Viewing array is NOT individual of original array
view_arr = org_arr.view()                                                          
org_arr[0] = -2                                                                    # Affects the 'viewed' array
print("==> org_arr: ",org_arr," view array:",view_arr)

print("Know whether an array is a copy or a view")
print("==> ",view_arr.base)                                                               # This returns the array with changes
print("==> ",copied_arr.base)                                                             # This returns None
print("==> ",org_arr.base)

heading("SHAPE AND RESHAPE ARRAYS")
print("Shape gives you the row X column size of an array. Bascially dimensionality")
print("==> Shape of threedimarr: ",threedimarr.shape)
print("==> Shape of fivedimarr: ",fivedimarr.shape)

print("==> Reshape an array. Change the dimensionality of an array. Ex: 1X6 array can be converted into 2X3 array")
sampleonedimarr = np.array([5,'False',-54,0.2,8,9.8])
print("==> sample one dim arr: ",sampleonedimarr)
reshapedintotwodimarr = sampleonedimarr.reshape(2,3)                              # 2d array contains 2 arrays inside it with 3 elements
                                                                                  # in each of the arrays
print("==> reshaped into two dim arr:",reshapedintotwodimarr)

reshapedintothreedimarr = sampleonedimarr.reshape(3,2,1)                          # This means the 3d array contains 3 arrays in it. In those 3 arrays
                                                                                  # there are two arrays with 1 element each
print("==> reshaped into three dim arr:", reshapedintothreedimarr)

heading("BASIC OPERATIONS ON ARRAYS")
                                                        # Create an array from the range a to b
first10nums = np.arange(1,11)                                                     # Gives array with numbers from 1 to 10 (11-1)
print("==> first 10 numbers",first10nums)

first10oddnums = np.arange(1,20,2)
print("==> first 10 odd numbers",first10oddnums)

sum = first10nums + first10oddnums
print("==> sum of two arrays",sum)

diff = first10nums - first10oddnums
print("==> diff of two arrays",diff)

prod = first10nums * first10oddnums
print("==> prod of two arrays",prod)

div = first10nums / first10oddnums
print("==> div of two arrays",div)

squareofoddnums = first10oddnums**2
print("==> square of 10 odd nums",squareofoddnums)

sqrootoffirst10nums = np.sqrt(first10nums)
print("==> square root of first 10 nums",sqrootoffirst10nums)



bmi_results = weight_array/ height_array**2                                    # ** operator to calculate exponent
print("==>",bmi_results, type(bmi_results))

                                                        ##Subsetting
print("==> Is each element greater than 25?",bmi_results>25)                                                        #  evaluates this boolean expression over entire numpy array

print("==> Printing values greater than 25",bmi_results[bmi_results>25])                                            # gives subset of numpy array with the condition applied

heading("DATA ANALYTICS OPERATIONS")
samplings = np.array([5,6])
print("==> sum of array:",samplings,np.sum(samplings))                                                              # same as 'samplings.sum()'
samplings2d = np.array([[23,3,10],[9,6,5]])            
print("sum of 2d array",samplings2d,np.sum(samplings2d))                                                            # same as 'samplings2d.sum()'

print("==> maximum of all elements in 1D array",samplings.max())                                                    # max() value of all elements in the array
print("==> minimum of all elements in 2D array",samplings2d.min())                                                  # min() value of all elements in the array
 
print("==> printing sum of all nums in each column",np.sum(samplings2d,axis = 0))                                   #sum value of all elements in the array - columnwise
print("==> printing sum of all nums in each row",np.sum(samplings2d,axis = 1))                                      #sum value of all elements in the array - rowwise

heading("ITERATIONS ON ARRAYS")
print("==> Printing 1d samplings array elements")
for ele in samplings:
    print(ele)
print("==> Printing 2d samplings2d array elements")
for ele in samplings2d:
    print(ele)

for ele in samplings2d:
    for num in ele:
        print(num)

print("==> Printing 2d samplings2d array elements using 'nditer'")
for ele in np.nditer(samplings2d):
    print(ele)

print("==> Printing 2d samplings2d array elements using 'ndenumerate'")
for ele in np.ndenumerate(samplings2d):
    i,e = ele[0],ele[1]
    print("ele:",e,"index: ",i)

heading("JOINING AND SPLITTING ARRAYS")
print("==> Concatenate two arrays")
joinedarrays = np.concatenate((height, weight_array))
print("==> ",joinedarrays)

print("==> Join arrays along rows")
joinedarraysbyrow = np.concatenate((samplings2d,twodimarr),axis = 1)                                              # No. of rows should be same for two arrrays when joining by row
                                                                                                                  # Similary no.of columns should be same for two arrays wnen joining by column
print(joinedarraysbyrow)
print("==> Splitting arrays")
splittedarray = np.array_split(onedimarr,5)                                                                       # There is no restriction in specifying number of segments. 
                                                                                                                  # If no.of segments > no of array elements, new empty arrays are added
print("==> Splitted array",splittedarray)
print("==> 3rd element in splitted array",splittedarray[2])

splittedarray2d = np.array_split(twodimarr,5)                                                                     # Splitting 2d array is done by splitting by 'row'
print("==> ",splittedarray2d)

loc = np.where(onedimarr == 92.98)
print("==> Loc",loc)

anotherloc = np.where(threedimarr == 'False')                                                                     # We can also use ==, >,<, >=,<= and many more

print("==> Another Loc",anotherloc)
oddnums = np.where(samplings %2 != 0 )
print("==> Oddnums",oddnums)

heading("SORTING ARRAYS")
sorted_onedimearr = np.sort(onedimarr) 
print("==> One dim arr:",onedimarr," sorted one dim array:",sorted_onedimearr)

sorted_samplings2d = np.sort(samplings2d)
print("==> ",sorted_samplings2d)                                                                                        # Sorts elements in each row 

sorted_twodimarr = np.sort(twodimarr)
print("==> twodimarr: ",twodimarr,"sorted two dim arr: ", sorted_twodimarr)

places = np.array(["hyderabad","Haiti","newyork","1loudon","Hawaii"])
print("==> ",np.sort(places))

print("==> Using searchsorted()")                                                                                # Returns the possible index of search key item incase
                                                                                                                 # if we were to place the key in the sorted array

print("==> Location of 7 in onedim sorted array: ",np.searchsorted(onedimarr,7))

print("==> Location of 0.2,-22.5,89.99 in onedim sorted array:", np.searchsorted(onedimarr,[0.2,-22.5,89.99]))
```
